### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y_line = [i**2 for i in x]
bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y = X^2')
axs[0, 0].grid(True)

# Bar chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot
axs[1, 0].scatter(x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
for i in range(len(x)):
    axs[1, 0].annotate(f'({x[i]}, {scatter_y[i]})', (x[i], scatter_y[i]), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].grid(True)

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio to make it a circle

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

```
